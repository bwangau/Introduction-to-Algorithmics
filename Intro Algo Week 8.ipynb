#### Example 1: A simple graph

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (3,1)])
```

```python
pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True)
plt.show()
```

#### Example 2: Friendship network - An undirected graph

```python
# Create the graph 
G = nx.Graph()
graph = {"Alice": ["Bob", "Charlie"],
    "Bob": ["Alice", "David"],
    "Charlie": ["Alice"],
    "David": ["Bob"]
}
```

```python
# Add edges to the graph 
for person, friends in graph.items():
    for friend in friends:
        G.add_edge(person, friend)
```

```python
# Print all the nodes and edges
print(G.nodes())
print(G.edges())
```

```python
# Draw the graph
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', font_weight='bold')
plt.title("Friendship Network")
plt.show()
```

#### Example 3: A directed graph 

```python
flights = [("Shanghai", "Osaka"), ("Osaka", "Busan"), ("Busan", "Shanghai")]
# Create a directed graph 
G = nx.DiGraph()
G.add_edges_from(flights)
# Draw with arrows
nx.draw(G, with_labels=True, node_size=1000, arrows=True)
plt.title("Flight Routes")
plt.show()
```
